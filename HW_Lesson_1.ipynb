__Задача 1__

Группа студентов изучает 10 различных дисциплин. Сколькими способами можно составить расписание на понедельник, если в этот день должно быть 4 разных занятия?

Это размещения, т.к. важен порядок следования занятий

```python
def arrangements(n: int, k: int) -> int:
    """
        Число размещений
    """
    return np.math.factorial(n) // np.math.factorial(n - k)

print(arrangements(10,4))
```

__Задача 2__

Из колоды в 52 карты вынимают случайным образом 4 карты. Найти число исходов, соответствующих тому, что был вытянут хотя бы один туз.

Один туз из 4-х, это сочетания и 3 оставшиеся карты (без туза) из 51 тоже сочетания. Общее число - перемножение обоих сочетаний 


```python
def combinations(n: int, k: int) -> int:
    """
        Число сочетаний
    """
    return np.math.factorial(n) // (np.math.factorial(k) * np.math.factorial(n-k))

print(combinations(4, 1) * combinations(51, 3))
```

Семь человек рассаживаются наудачу на скамейке. Какова вероятность того, что два определённых человека будут сидеть рядом?

Здесь сочетания, т.к. нам не важно как Вася и Петя сядут друг относительно друга, важно, чтобы они сели рядом. Выбор у них будет не из 7 мест, а из 6-ти, т.к. они занимают 2 места. Следовательно, ищем число сочетаний из 6 по 2

```python
print(combinations(6, 2))
```

__Задача 4__

Из 60 вопросов, входящих в экзаменационные билеты, студент знает 50. Какова вероятность того, что среди трёх наугад выбранных вопросов студент знает: а) все? б) два?

Для расчета вероятности нужно поделить число искомых вариантов (все верные вытянул, например) на общее кол-во возможных вариантов.
Общее число вариантов - сочетания из 60 по 3
Все верные - сочетания из 50 по 3
Два верных  - сочетания из 50 по 2, умноженное на сочетания из оставшихся 10 (которые он не знает) по 1


```python
print("Вероятность вытянуть 3 выученных билета:", combinations(50,3)/combinations(60,3))
print("Вероятность вытянуть 2 выученных билета:", (combinations(50,2)*combinations(10, 1))/combinations(60,3))
```
